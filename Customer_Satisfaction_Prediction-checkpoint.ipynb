```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
import nltk
from nltk.corpus import stopwords
from collections import Counter

# --- Sci-kit learn and Imbalanced-learn imports ---
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.feature_extraction.text import TfidfVectorizer

# Try to import imblearn for SMOTE
try:
    from imblearn.over_sampling import SMOTE
except ImportError:
    print("imbalanced-learn not installed. Please install with: pip install imbalanced-learn")
    SMOTE = None

# Try to import XGBoost
try:
    from xgboost import XGBClassifier
except ImportError:
    print("XGBoost not installed. Skipping XGBoost model.")
    XGBClassifier = None

# Try to import WordCloud (optional for visualizations)
try:
    from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
except ImportError:
    print("WordCloud not installed. Skipping WordCloud visualizations.")
    WordCloud = None

# For saving/loading models
try:
    import joblib
except ImportError:
    print("joblib not installed. Please install with: pip install joblib")
    joblib = None

# --- NLTK Downloads (run once) ---

nltk.download('stopwords')
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# --- GLOBAL MAPPINGS / HELPERS ---
# Define mapping from original numeric Customer Satisfaction Rating to 'Low', 'Medium', 'High' categories
numeric_to_categorical_mapping = {
    1.0: 'Low',
    2.0: 'Low',
    3.0: 'Medium',
    4.0: 'High',
    5.0: 'High'
}
# Define mapping from 'Low', 'Medium', 'High' categories to integers for model training
target_encoding_mapping = {'Low': 0, 'Medium': 1, 'High': 2}


# Helper Functions (defined at the top for clarity)
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'<.*?>', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def remove_stopword(x, custom_stopwords=None):
    if custom_stopwords is None:
        custom_stopwords = []
    stpwrd = set(nltk.corpus.stopwords.words('english'))
    stpwrd.update(custom_stopwords)
    return [y for y in x if y not in stpwrd]
```

```python
# --- 1. Data Loading ---
print("--- Data Loading ---")
customers = pd.read_csv("customer_support_tickets.csv")
print(f"Initial customers DataFrame shape: {customers.shape}")
```

```python
# --- 2. Initial Data Cleaning & Feature Engineering for `customers` DataFrame ---
print("\n--- Initial Data Cleaning & Feature Engineering on 'customers' DataFrame ---")

# Convert to datetime
customers['Date of Purchase'] = pd.to_datetime(customers['Date of Purchase'], format='%Y-%m-%d', errors='coerce')
customers['First Response Time'] = pd.to_datetime(customers['First Response Time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
customers['Time to Resolution'] = pd.to_datetime(customers['Time to Resolution'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Create `Resolution Time` (Timedelta)
customers['Resolution Time (Timedelta)'] = customers['Time to Resolution'] - customers['First Response Time']
customers['Resolution Time (Timedelta)'] = customers['Resolution Time (Timedelta)'].apply(lambda x: x if pd.notna(x) and x >= pd.Timedelta(0) else pd.NaT)

# Convert to numeric hours (will impute NaNs later on df_resolution subset)
customers['Interval_Hours'] = customers['Resolution Time (Timedelta)'].dt.total_seconds() / 3600

# Create 'Has_Response' and 'Has_Resolution' flags based on recorded times
customers['Has_Response'] = customers['First Response Time'].notna().astype(int)
customers['Has_Resolution'] = customers['Time to Resolution'].notna().astype(int)

# Derive 'Type of Customer' (age groups)
age_group = []
for i in customers['Customer Age']:
    if i <= 30:
        age_group.append('Young Customer')
    elif 30 < i < 55:
        age_group.append('Middle Age Customer')
    else:
        age_group.append('Old Customer')
customers['Type of Customer'] = age_group

# Clean Ticket Description (Applies to all customers, for TF-IDF and Sentiment)
customers['Ticket Description Cleaned'] = customers['Ticket Description'].apply(lambda x: clean_text(x))

# APPLY VADER SENTIMENT FEATURES HERE, ON CUSTOMERS DF
sid = SentimentIntensityAnalyzer()
customers['Sentiment_Description_Compound'] = customers['Ticket Description Cleaned'].apply(lambda x: sid.polarity_scores(x)['compound'])
customers['Sentiment_Description_Positive'] = customers['Ticket Description Cleaned'].apply(lambda x: sid.polarity_scores(x)['pos'])
customers['Sentiment_Description_Negative'] = customers['Ticket Description Cleaned'].apply(lambda x: sid.polarity_scores(x)['neg'])
customers['Sentiment_Description_Neutral'] = customers['Ticket Description Cleaned'].apply(lambda x: sid.polarity_scores(x)['neu'])

# Drop 'Ticket ID'
customers = customers.drop(columns=['Ticket ID'], errors='ignore')

print(f"Customers DataFrame shape after initial feature engineering: {customers.shape}")


# --- Initialize and Fit TF-IDF Vectorizer Here ---
print("\n--- Initializing and Fitting TF-IDF Vectorizer (on all cleaned descriptions) ---")
tfidf_vectorizer = TfidfVectorizer(max_features=2000, stop_words='english', min_df=5, max_df=0.8)
tfidf_vectorizer.fit(customers['Ticket Description Cleaned'])
print("TF-IDF Vectorizer fitted.")
```

```python
# --- 3. Filter for Resolved Tickets and Define Target Variable (`df_resolution`) ---
print("\n--- Filtering for Resolved Tickets & Defining Target Variable ('df_resolution') ---")
print(f"Shape before filtering for 'Has_Resolution': {customers.shape}")
print(f"Value counts of 'Has_Resolution' in 'customers':\n{customers['Has_Resolution'].value_counts()}")

df_resolution = customers[customers['Has_Resolution'] == 1].copy()
print(f"Shape after filtering for 'Has_Resolution' == 1: {df_resolution.shape}")

# Convert 'Customer Satisfaction Rating' to numeric
df_resolution['Customer Satisfaction Rating'] = pd.to_numeric(df_resolution['Customer Satisfaction Rating'], errors='coerce')
print(f"Value counts of 'Customer Satisfaction Rating' BEFORE NaN drop (after filtering Has_Resolution):\n{df_resolution['Customer Satisfaction Rating'].value_counts(dropna=False)}")

df_resolution.dropna(subset=['Customer Satisfaction Rating'], inplace=True)
print(f"Shape after dropping NaNs in 'Customer Satisfaction Rating': {df_resolution.shape}")

# Define the target 'Type of rating' using `numeric_to_categorical_mapping`
df_resolution['Type of rating'] = df_resolution['Customer Satisfaction Rating'].map(numeric_to_categorical_mapping)
print(f"Value counts of 'Type of rating' BEFORE final NaN drop:\n{df_resolution['Type of rating'].value_counts(dropna=False)}")

df_resolution.dropna(subset=['Type of rating'], inplace=True) # This should not drop rows now if mapping is correct
print(f"Shape after dropping NaNs in 'Type of rating': {df_resolution.shape}")

# Fill NaN in `Interval_Hours` ONLY for the `df_resolution` subset
if 'Interval_Hours' in df_resolution.columns and df_resolution['Interval_Hours'].isnull().any():
    median_interval_hours_resolved = df_resolution['Interval_Hours'].median()
    df_resolution['Interval_Hours'] = df_resolution['Interval_Hours'].fillna(median_interval_hours_resolved) # Avoid inplace warning
    print(f"Interval_Hours NaNs filled in df_resolution. Remaining NaNs: {df_resolution['Interval_Hours'].isnull().sum()}")


print("df_resolution head after final preprocessing steps for its construction:")
print(df_resolution.head())
print(f"Final shape of df_resolution: {df_resolution.shape}")
if df_resolution.empty:
    raise ValueError("df_resolution is EMPTY after filtering and dropping NaNs. Cannot proceed. Check data or preprocessing logic. Values from Customer Satisfaction Rating may not match keys in numeric_to_categorical_mapping.")
```

```python
# --- 5. Prepare `X` (Features) and `y` (Target) for Modeling Pipeline ---
print("\n--- Constructing Features (X) and Target (y) (Cont. from EDA) ---")

feature_columns = [
    'Customer Age', 'Customer Gender', 'Product Purchased', 'Ticket Type',
    'Ticket Channel', 'Ticket Priority', 'Ticket Subject', 'Type of Customer',
    'Has_Response', 'Has_Resolution', 'Interval_Hours', # Flags and numeric interval
    'Sentiment_Description_Compound', 'Sentiment_Description_Positive',
    'Sentiment_Description_Negative', 'Sentiment_Description_Neutral' # Sentiment scores
]

# Create X_base from df_resolution for these columns
X_base = df_resolution[feature_columns].copy()

# TRANSFORM TF-IDF features for the `df_resolution` subset
tfidf_features_sparse_resolved = tfidf_vectorizer.transform(df_resolution['Ticket Description Cleaned'])
tfidf_df_resolved = pd.DataFrame(tfidf_features_sparse_resolved.toarray(),
                                 columns=['tfidf_' + col for col in tfidf_vectorizer.get_feature_names_out()],
                                 index=df_resolution.index)

X = pd.concat([X_base, tfidf_df_resolved], axis=1)

y = df_resolution['Type of rating'].copy()
y_encoded = y.map(target_encoding_mapping)


print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y_encoded.shape}")
print("Sample X columns and their dtypes after construction (before final encoding/scaling):")
print(X.info(memory_usage="deep"))
```

```python
# --- 6. Categorical Encoding (LabelEncoder) ---
print("\n--- Applying Categorical Encoding ---")
categorical_maps = {} # Reset or initialize as global variable

cat_cols_for_encoding_in_X = X.select_dtypes(include='object').columns.tolist()

for col in cat_cols_for_encoding_in_X:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    categorical_maps[col] = dict(zip(le.classes_, le.transform(le.classes_)))

print("X head after Label Encoding:")
print(X.head())
```

```python
# --- 7. Feature Scaling (StandardScaler) ---
print("\n--- Applying Feature Scaling ---")
scaler_for_deployment = StandardScaler()
scaler_for_deployment.fit(X)

X_scaled = scaler_for_deployment.transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)


print("X_scaled head after StandardScaler:")
print(X_scaled.head())
```

```python
# --- 8. Train-Test Split (from scaled data) ---
X_train_raw, X_test, y_train_raw, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print(f"\nRaw training set shape before selection: {X_train_raw.shape}")
print(f"Raw training target distribution: \n{y_train_raw.value_counts()}")
```

```python
# --- 9. Feature Selection (SelectKBest) ---
print("\n--- Applying Feature Selection (SelectKBest) ---")
selector_for_deployment = SelectKBest(score_func=chi2, k=min(1000, X_train_raw.shape[1]))
selector_for_deployment.fit(X_train_raw.abs(), y_train_raw)

selected_features_names = X_train_raw.columns[selector_for_deployment.get_support()].tolist()
print(f"Selected {len(selected_features_names)} features for modeling.")

X_train_selected = X_train_raw[selected_features_names]
X_test_selected = X_test[selected_features_names]
```

```python
# --- 10. No SMOTE - Train Directly on Imbalanced Data ---
print("\n--- Training on Imbalanced Data (SMOTE Skipped as requested) ---")
# Directly use X_train_selected and y_train_raw for final training data
X_train_final = X_train_selected
y_train_final = y_train_raw

print(f"Final training set shape: {X_train_final.shape}")
print(f"Final training target distribution: \n{y_train_final.value_counts()}")
```

```python
# --- 11. Model Training and Evaluation ---
print("\n--- Model Training and Evaluation ---")
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=2000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
}
if XGBClassifier:
    models["XGBoost"] = XGBClassifier(objective='multi:softmax', eval_metric='mlogloss', use_label_encoder=False, random_state=42)

best_model_overall = None
best_accuracy_overall = -1.0

for name, model in models.items():
    print(f"\n--- Training {name} ---")
    model.fit(X_train_final, y_train_final)
    y_pred = model.predict(X_test_selected)

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for {name}: {accuracy:.4f}")
    print(f"Classification Report for {name}:\n{classification_report(y_test, y_pred, target_names=['Low', 'Medium', 'High'])}")

    if accuracy > best_accuracy_overall:
        best_accuracy_overall = accuracy
        best_model_overall = model

print(f"\nBest performing model (based on test set accuracy): {type(best_model_overall).__name__} with accuracy {best_accuracy_overall:.4f}")
```

```python
# --- Hyperparameter Tuning  ---
if isinstance(best_model_overall, RandomForestClassifier) or (not isinstance(best_model_overall, (LogisticRegression, GradientBoostingClassifier, XGBClassifier)) and best_model_overall is not None):
    print("\n--- Hyperparameter Tuning with GridSearchCV for Random Forest ---")
    param_grid_rf = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2]
    }
    grid_search_rf = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid_rf,
                                  cv=3, scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search_rf.fit(X_train_final, y_train_final)

    tuned_rf_model = grid_search_rf.best_estimator_
    y_pred_tuned_rf = tuned_rf_model.predict(X_test_selected)
    tuned_accuracy = accuracy_score(y_test, y_pred_tuned_rf)

    print(f"\nBest parameters for Tuned Random Forest: {grid_search_rf.best_params_}")
    print(f"Best score (GridSearchCV) for Tuned Random Forest: {grid_search_rf.best_score_:.4f}")
    print(f"Accuracy of Tuned Random Forest on test set: {tuned_accuracy:.4f}")
    print(f"Classification Report for Tuned Random Forest:\n{classification_report(y_test, y_pred_tuned_rf, target_names=['Low', 'Medium', 'High'])}")

    if tuned_accuracy > best_accuracy_overall:
        best_model_overall = tuned_rf_model
        best_accuracy_overall = tuned_accuracy
    else:
        print("Tuned Random Forest did not surpass initial best model. Keeping original best model.")
```

```python
# --- 11. Feature Importance (from the Final Best Model) ---
print("\n--- Feature Importance from the Final Best Model ---")
if hasattr(best_model_overall, 'feature_importances_'):
    feature_importances = pd.Series(best_model_overall.feature_importances_, index=X_train_final.columns)
    feature_importances_sorted = feature_importances.sort_values(ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x=feature_importances_sorted.head(20).values, y=feature_importances_sorted.head(20).index, palette='viridis')
    plt.title('Top 20 Feature Importances (Final Best Model)', fontsize=16)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.tight_layout()
    plt.show()
else:
    print("The final best model does not have feature_importances_ attribute.")
```

```python
# --- 12. Save All Necessary Artifacts for Deployment ---
print("\n--- Saving All Necessary Artifacts for Deployment ---")

import os
artifacts_dir = 'streamlit_artifacts'
if not os.path.exists(artifacts_dir):
    os.makedirs(artifacts_dir)
    print(f"Created directory: {artifacts_dir}")


if joblib is None:
    print("Joblib not imported. Cannot save models. Please install joblib.")
else:
    joblib.dump(best_model_overall, os.path.join(artifacts_dir, 'best_customer_satisfaction_model.joblib'))
    joblib.dump(tfidf_vectorizer, os.path.join(artifacts_dir, 'tfidf_vectorizer.joblib'))
    joblib.dump(scaler_for_deployment, os.path.join(artifacts_dir, 'scaler.joblib'))
    joblib.dump(selector_for_deployment, os.path.join(artifacts_dir, 'selector.joblib'))
    joblib.dump(selected_features_names, os.path.join(artifacts_dir, 'selected_features.joblib'))
    joblib.dump(categorical_maps, os.path.join(artifacts_dir, 'categorical_maps.joblib'))
    joblib.dump(X.columns.tolist(), os.path.join(artifacts_dir, 'scaler_fit_column_names.joblib'))


print("\n--- All necessary artifacts successfully saved for Streamlit deployment! ---")
plt.close('all') # Close all matplotlib figures
```
